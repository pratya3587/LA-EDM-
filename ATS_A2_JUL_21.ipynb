```python
def MID(s, start, length):
    # example: MID("apple",2,3) returns "ppl"
    return str(s[start - 1: start - 1 + length])

import glob
import pandas as pd

df_list = []
for file in glob.glob(r'C:\Users\prati\Downloads\2021\July\*.csv'):
    # Tweak this to work for your actual filepaths, if needed.
   
    course_id = MID(file,47,10)
    #print(course_id)
    df = (pd.read_csv(file)
            .assign(Course_id=course_id))
    df_list.append(df)

df = pd.concat(df_list)
```

```python
table = pd.pivot_table(df,index=['user_id'],values=['Course_id'], aggfunc=pd.Series.nunique)
table.info()
```

```python
table=table.drop(table[table['Course_id']<2].index,axis=0)
table.info()
```

```python
df_1 = pd.merge(df, table, on='user_id', how='inner')
df_1.head()
```
