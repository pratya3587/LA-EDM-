Q1) Define a 1D array A containing natural numbers from 1 to 100.
1. Reshape the array into a 2D array R of size 10x10.
2. Create a binary matrix B of size 10x10 containing 1 at the position where there is a prime
in B and zero elsewhere.
3. Concatenate the above 2 matrices.
Note- A binary matrix is one that contains only zero and one


```python
!pip install numpy
import numpy as np 
```

```python
# 1D array 

A = np.array(range(1,101))
print(A)



```

```python
R = np.reshape(A,(10, 10))
print(R)
```

```python
B = np.zeros(100, dtype = int) 
B
```

```python
for num in range(1,101):  
   if num > 1:  
       for i in range(2,num):  
           if (num % i) == 0:  
               break  
       else:  
           B[i]=1
B[1]=1           
```

```python
B
```

```python
D = np.reshape(B,(10, 10))
D
```

```python

np.concatenate((R,D),axis=1)
```
